```python
# Importing important files 

import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 

print('Modules are imported.')
```

```python
# Importing COVID 19 Dataset 

corona_dataset_csv = pd.read_csv('covid19_Confirmed_dataset.csv')
corona_dataset_csv.head(10)
```

```python
# Finding number or rows x columns 

corona_dataset_csv.shape
```

```python
# Deleting columnc not necessary for our analysis 

corona_dataset_csv.drop(['Lat','Long'],axis=1,inplace=True)
corona_dataset_csv.head(10)
```

```python
# Aggregating rows by country 

corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
corona_dataset_aggregated.head(10)
```

```python
# Find maximum infection rate for all countries 

# Forming a list of all the countries 
countries = list(corona_dataset_aggregated.index)

# Creat an empty list - maximum infections and a for loop
max_infection_rates = []
for country in countries :                                                        # For each country, calculate the derivative and then max rate
    max_infection_rates.append(corona_dataset_aggregated.loc[country].diff().max())      # Directly append the nos. of max rate to created list 
corona_dataset_aggregated['max infection rate'] = max_infection_rates                 # Creating a new column - maximum infections 

corona_dataset_aggregated.head()
```

```python
# we only need max infection rate data, defining a new dataframe 

corona_data = pd.DataFrame(corona_dataset_aggregated['max infection rate'])
corona_data.head()
```

```python
# Importing world happiness report 

world_happiness_report = pd.read_csv("worldwide_happiness_report.csv")
world_happiness_report.head()
```

```python
# Dropping columns not important to our analysis 

columns_to_dropped = ['Overall rank','Score','Generosity','Perceptions of corruption']
world_happiness_report.drop(columns_to_dropped,axis=1 , inplace=True)

# Changing indices of data frame 
world_happiness_report.set_index(['Country or region'],inplace=True)

world_happiness_report.head()
```

```python
# Joining both datasets 

# Checking corona dataset  

corona_data.head()
```

```python
corona_data.shape
```

```python
world_happiness_report.head()
```

```python
world_happiness_report.shape
```

```python
# Both data don't have equal rows or countries, so use inner joint 

#The INNER JOIN creates a new result table by combining column values of two tables (table1 and table2) based upon the join-predicate. 
#The query compares each row of table1 with each row of table2 to find all pairs of rows which satisfy the join-predicate.

data = world_happiness_report.join(corona_data,how="inner")
data.head()
```

```python
# Correlation between development index and infection rate 

data.corr()
```
